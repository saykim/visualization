```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.set(style='ticks')
```

```python
tips = sns.load_dataset('tips')
```

```python
g = sns.FacetGrid(tips, col='time')
```

```python
g = sns.FacetGrid(tips, col='time')
g.map(plt.hist, 'tip')
```

```python
g = sns.FacetGrid(tips, col='sex', hue='smoker')
g.map(plt.scatter, 'total_bill', 'tip', alpha=)
```
