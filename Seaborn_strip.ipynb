```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
plt.style.use('seaborn')
```

```python
sns.stripplot(x='tip', data=tips)
```

```python
sns.stripplot(x='tip', data=tips, color='green')
```

```python
sns.stripplot(x=tips['total_bill'])
```

```python
sns.stripplot(x='day', y='total_bill', data=tips)
```

```python
sns.stripplot(x='day', y='total_bill', data=tips, jitter=False)
```

```python
sns.stripplot(x='day', y='total_bill', data=tips, jitter=0.1)
```

```python
sns.stripplot(x='total_bill', y='day', data=tips)
```

```python
sns.stripplot(x='total_bill', y='day', data=tips, linewidth=1.2)
```

```python
sns.stripplot(x='day', y='total_bill', data = tips, hue='sex', size=4)
```

```python
sns.stripplot(x='day', y='total_bill', data = tips, hue='sex', \
              size=3, dodge=True, palette='winter_r') #hue dodge  
```

```python
sns.stripplot(x='day', y='total_bill', data = tips, hue='sex', \
              size=3, dodge=True, palette='winter_r', order=['Fri', 'Sun', ' Sat', 'Thur']) #X  
```
