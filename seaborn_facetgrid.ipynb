```python
import numpy as np, pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.FacetGrid(tips, row='smoker', col='time')
```

```python
x = sns.FacetGrid(tips, row='smoker', col='time')
x = x.map(plt.hist, 'total_bill')
```

```python
x = sns.FacetGrid(tips, row='smoker', col='time', hue='sex')
x = x.map(plt.hist, 'total_bill')
```

```python
x = sns.FacetGrid(tips, row='smoker', col='time', hue='sex')
x = x.map(plt.hist, 'total_bill', bins=15, alpha=0.7)
```

```python
g = sns.FacetGrid(tips, col='time', row='smoker')
g = g.map(plt.scatter, 'total_bill', 'tip')
```

```python
g = sns.FacetGrid(tips, col='time', row='smoker', hue='sex')
g = g.map(plt.scatter, 'total_bill', 'tip', alpha=0.5)
```

```python
g = sns.FacetGrid(tips, col='time', row='smoker')
g = g.map(sns.regplot, 'total_bill', 'tip')
```
